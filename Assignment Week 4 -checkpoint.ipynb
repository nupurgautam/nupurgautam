```python
# cleaning the data set
# Differentiate numerical features (minus the target) and categorical features
y_train = train_df['SalePrice']
train_df.drop(['SalePrice'],axis =1,inplace=True)
len(train_df.columns)
len(test_df.columns)
```

```python
categorical_features_tr = train_df.select_dtypes(include = ["object"]).columns
numerical_features_tr = train_df.select_dtypes(exclude = ["object"]).columns
print("Numerical features : " + str(len(numerical_features_tr)))
print("Categorical features : " + str(len(categorical_features_tr)))
train_num = train_df[numerical_features_tr]
train_cat = train_df[categorical_features_tr]
test_num = test_df[numerical_features_tr]
test_cat = test_df[categorical_features_tr]
```

```python
# Handle remaining missing values for numerical features by using median as replacement
print("NAs for numerical features in train : " + str(train_num.isnull().values.sum()))
train_num = train_num.groupby(train_num.columns, axis = 1).transform(lambda x: x.fillna(x.median()))
test_num = test_num.groupby(test_num.columns, axis = 1).transform(lambda x: x.fillna(x.median()))
print("NAs for numerical features in train : " + str(train_num.isnull().values.sum()))
print("Remaining NAs for numerical features in train : " + str(train_num.isnull().values.sum()))
train_cat = pd.get_dummies(train_cat)
test_cat = pd.get_dummies(test_cat)
```

```python
missing_cols = set(train_cat.columns) - set(test_cat.columns)
len(missing_cols)
for c in missing_cols:
        test_cat[c] = 0
```

```python
print("Remaining NAs for categorical features in train : " + str(train_cat.isnull().values.sum()))
print("Remaining NAs for categorical features in test : " + str(test_cat.isnull().values.sum()))
```

```python
# joining the two sets
train = pd.concat([train_cat,train_num],axis=1)
test = pd.concat([test_cat,test_num],axis=1)
```

```python
X_train, X_train_test, y_train, y_train_test = train_test_split(train, y_train, random_state=0,test_size=0.3)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_train_test shape: {}".format(X_train_test.shape))
print("y_train_test shape: {}".format(y_train_test.shape))
```

```python
n_folds = 5
```

```python
scorer = make_scorer(mean_squared_error,greater_is_better = False)
def rmse_CV(model,y_i,x_i):
    kf = KFold(n_folds,shuffle=True,random_state=42).get_n_splits(train.values)
    rmse = np.sqrt(-cross_val_score(model,X=x_i,y=y_i,scoring ="neg_mean_squared_error",cv=kf))
    return (rmse)
```
